### Fitting data in various models

```python
## import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pickle
from sklearn.utils import resample
```

```python
## SVM classifier
logging.info('Importing data in svm_classifier model')
def svm_classifier(X_train_selected,X_test_selected,y2_train,y2_test):
    classifier_svm=SVC(kernel='rbf',random_state=0)
    classifier_svm.fit(X_train_selected,y2_train)
    y_pred=classifier_svm.predict(X_test_selected)
    cm=confusion_matrix(y2_test,y_pred)
    return (f'Train Score:{classifier_svm.score(X_train_selected,y2_train)}\n Test Score:{classifier_svm.score(X_test_selected,y2_test)}')
```

```python
## KNN classifier model
logging.info('Import data in KNN_classifier model')
def knn_classifier(X_train_selected,X_test_selected,y2_train,y2_test):
    classifier_knn=KNeighborsClassifier(metric='minkowski',p=2)
    classifier_knn.fit(X_train_selected,y2_train)
    y_pred=classifier_knn.predict(X_test_selected)
    cm=confusion_matrix(y2_test,y_pred)
    return (f'Train Score:{classifier_knn.score(X_train_selected,y2_train)}\n Test Score:{classifier_knn.score(X_test_selected,y2_test)}')
```

```python
## DecisionTree classifier model
logging.info('import data in Decision Tree classifier model')
def tree_classifier(X_train_selected,X_test_selected,y2_train,y2_test):
    classifier_tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
    classifier_tree.fit(X_train_selected,y2_train)
    y_pred=classifier_tree.predict(X_test_selected)
    cm=confusion_matrix(y2_test,y_pred)
    return (f'Train Score:{classifier_tree.score(X_train_selected,y2_train)}\n Test Score:{classifier_tree.score(X_test_selected,y2_test)}')
```

```python
## RandomForest classifier model
logging.info('Import data in RandomForest classifer model')
def forest_classifier(X_train_selected,X_test_selected,y2_train,y2_test):
    classifier_forest=RandomForestClassifier(criterion='entropy',random_state=0)
    classifier_forest.fit(X_train_selected,y2_train)
    y_pred=classifier_forest.predict(X_test_selected)
    cm=confusion_matrix(y2_test,y_pred)
    
    return (f'Train Score:{classifier_forest.score(X_train_selected,y2_train)}\n Test Score:{classifier_forest.score(X_test_selected,y2_test)}')
    
```

```python
logging.info('Now print the score of each and every model')
def print_score(X_train_selected, X_test_selected, y2_train, y2_test):
    print("SVM:\n")
    result1=svm_classifier(X_train_selected, X_test_selected, y2_train, y2_test)
    print(result1)
    print("-"*100)
    print()

    print("KNN:\n")
    result2=knn_classifier(X_train_selected, X_test_selected, y2_train, y2_test)
    print(result2)

    print("-"*100)
    print()

    print("Decision Tree:\n")
    result4=tree_classifier(X_train_selected, X_test_selected, y2_train, y2_test)
    print(result4)

    print("-"*100)
    print()

    print("Random Forest:\n")
    result5=forest_classifier(X_train_selected, X_test_selected, y2_train, y2_test)
    print(result5)
```

```python
print_score(X_train_selected, X_test_selected, y2_train, y2_test)
```

```python
##Performance Metrics
## Performance Metrics
classifier_forest = RandomForestClassifier(criterion = 'entropy')
classifier_forest.fit(X_train_selected,y2_train)
y_pred = classifier_forest.predict(X_test_selected)
cm = confusion_matrix(y2_test,y_pred)
cm
```

```python
## Classification Report (Accuracy, Precision, Recall and F1 Score)
from sklearn.metrics import roc_auc_score,roc_curve,classification_report
```
